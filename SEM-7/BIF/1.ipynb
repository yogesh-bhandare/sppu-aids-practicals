```python
from Bio.Seq import Seq
from Bio.SeqUtils import gc_fraction


dna_sequence = "ATGTCATGATATAA"

seq=Seq(dna_sequence)

def motif(sequ,motif):
    return([i for i in range (len(sequ)) if sequ.startswith(motif,i)])
m=motif(seq,"ATA")
print(m)

print(gc_fraction(seq))

def Cd(dna_sequence):
    start_codon = "ATG"
    stop_codons = {"TAA", "TAG", "TGA"}  
    coding_regions = []

    i=0
    while i<len(dna_sequence):
        if dna_sequence.startswith(start_codon,i):
            start_index=i
            i+=3
            while i<len(dna_sequence):
                if dna_sequence[i:i+3] in stop_codons:
                    coding_regions.append((start_index,i+3))
                    break
                i+=3
        else:
            i+=1
    return coding_regions

print(Cd(dna_sequence))
                    
```
