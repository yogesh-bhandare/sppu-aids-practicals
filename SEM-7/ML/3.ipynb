```python
import matplotlib.pyplot as plt
from sklearn import datasets, svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# 1. load the dataset
digits = datasets.load_digits()
```

```python
# 2. prepare data for svm
# images are 8x8, data is a flattened 64-element array for each image
x = digits.data
y = digits.target
```

```python
# 3. split data into training and testing sets
# 70% for training, 30% for testing (test_size=0.3)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# 4. create and train the svm model
# using a support vector classifier (svc) with a radial basis function (rbf) kernel
# gamma is a kernel parameter, c is the regularization parameter
model = svm.SVC(gamma=0.001, C=100)
model.fit(x_train, y_train)
```

```python
# 5. make predictions
y_pred = model.predict(x_test)
```

```python
# 6. evaluate the model
accuracy = accuracy_score(y_test, y_pred)
confusion_matrix = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
print(f"svm classifier accuracy: {accuracy:.4f}\n")
print("confusion matrix:\n", confusion_matrix)
print("classification report:\n", report)
```
